### **Ejercicio 2 (C = 1.0)**

Genere la siguiente imagen [resultado](https://drive.google.com/file/d/13z0tvGYD-37_PuHOqoRWHGGmGq7xxvU7/view?usp=drive_link).

```python
# Lea la imagen
image = np.zeros((400,400))

space = 25

# Realice aquí todas las operaciones que necesite. Puede añadir cuantas celdas de código requiera
for i in range(400):
    for j in range(400):
        if (i + j) % (2 * space) < space:
            image[i, j] = 255
        else:
            image[i, j] = 0

image = np.flip(image, 0)


# Muestre la imagen y guardela en drive usando opencv (consultar)
plt.imshow(image, cmap='gray', vmin=0, vmax=1)

#guardar la imagen en el drive
cv2.imwrite(carpeta_raiz+"imagen_2.png", image)
```

### **Ejercicio 3 (C = 1.6)**

Genere la siguiente animación [resultado](https://drive.google.com/file/d/1jSNhV2CYdm6-E5wuo7o5wE-b2J-_mEil/view?usp=drive_link).

```python
# Lea la imagen

# Crear matriz de ceros (imagen inicial)
Ruta1 = carpeta_raiz+'bengal_tiger.jpg' #Ubicación de la imagen desde el google drive
image1 = plt.imread(Ruta1)
Ruta2 = carpeta_raiz+'lena.jpg' #Ubicación de la imagen desde el google drive
image2 = plt.imread(Ruta2)
# Crear figura y ejes
import numpy as np
import matplotlib.pyplot as plt

# Asegurarse de que las imágenes estén en el rango correcto y tengan dimensiones compatibles
image1_resized=image1[0:216, 0:250]
image2_resized=image2[0:216, 50:300]

# Crear figura y ejes
fig, ax = plt.subplots()

# Inicializar la imagen con la primera imagen
img = ax.imshow(image1_resized)

# Desvanecer la primera imagen y mostrar la segunda
frames = []
for alpha in np.linspace(0, 1, 8):
    combined_image = (alpha * image1_resized + (1 - alpha) * image2_resized).astype(np.uint8)
    img.set_array(combined_image)
    # Forzar la actualización de la visualización en cada iteración
    # Añade un pequeño retraso para visualizar la animación
    frames.append(combined_image)
plt.close(fig)
# Crear el archivo GIF
output_path = carpeta_raiz+'transicion.gif'
imageio.mimsave(output_path, frames, format='GIF', duration=280,loop=0)

print(f'El archivo GIF se ha creado en: {output_path}')
# Ruta del archivo GIF en Google Drive
ruta_gif = carpeta_raiz+'transicion.gif'
# Mostrar el GIF en Google Colab
Image(filename=ruta_gif)
```

### **Ejercicio 4 (C = 1.0)**

Genere la siguiente imagen [resultado](https://drive.google.com/file/d/1-9jWSwRb1ld_xN6iHTHVKT92bGat0I4s/view?usp=sharing).

```python
# Lea la imagen
image = cv2.imread(carpeta_raiz+'lena.jpg')

# Realice aquí todas las operaciones que necesite. Puede añadir cuantas celdas de código requiera
colors = -1
for i in range(0,300,50):
    for j in range(0,700,50):
        if colors > 2:
            colors = 0
        image[i:i+50, j-50:j, colors] = 255
        colors += 1
# Muestre la imagen y guardela en drive usando opencv (consultar)
plt.imshow(image, vmin=0, vmax=255)

#guardar la imagen en el drive
cv2.imwrite(carpeta_raiz+"imagen_4.png", image)
```

Genere la siguiente imagen [resultado](https://drive.google.com/file/d/10jxoa_muSn6bxsDBy7Vw88wz1oirMiRH/view?usp=drive_link)

```python
# Lea la imagen
image = cv2.imread(carpeta_raiz+'lena.jpg')

# Realice aquí todas las operaciones que necesite. Puede añadir cuantas celdas de código requiera
img_recortada = image[:150, :, :]
img_invertida = np.flip(img_recortada, 0)

image = np.zeros((300,400,3)).astype(np.uint8)
image[:150, :, :] = img_recortada
image[150:, :, :] = img_invertida

# Muestre la imagen y guardela en drive usando opencv (consultar)
plt.imshow(image[:,:, [2,1,0]], vmin=0, vmax=255)

#guardar la imagen en el drive
cv2.imwrite(carpeta_raiz+"imagen_5.png", image)
```
