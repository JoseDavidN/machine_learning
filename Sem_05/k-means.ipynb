# K Means - Generated data

Isotropic dataset module in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html) \
Train-test splits in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) \
k-Means module in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

```
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

```
X, _ = make_blobs(n_samples = 1000 , random_state = 170 )

# Create train (80%) and test (20%) sets and print all shapes
X_train, X_test = train_test_split(X, test_size=0.20, random_state=1)

_, axes = plt.subplots(1,2, figsize=(13,5))
axes[0].scatter( X[:, 0] , X[:, 1] , edgecolors='k', cmap=plt.cm.Paired )
axes[1].scatter( X_train[:, 0], X_train[:, 1], c='r', edgecolors='k' )
axes[1].scatter( X_test[:, 0], X_test[:, 1] , c='b', edgecolors='k' )
```

```
kmeans = KMeans(n_clusters=3, random_state=1)

# Train Kmeans
kmeans.fit(X_train)
```

```
# Plot centroids
centroids = kmeans.cluster_centers_
print(centroids)
```

```
# Create predictions
predictions = kmeans.predict(X_test)
```

```
_, axes = plt.subplots (1, 2, figsize = (10,5))
colors=['blue','cyan','magenta']

# plot data and found centroids
axes[0].scatter(X[:, 0], X[:, 1], c='g', edgecolors='k')
axes[0].scatter(centroids[:, 0], centroids[:, 1], c=colors, marker='*', s=500, edgecolors='k')
axes[1].scatter(X_test[:, 0], X_test[:, 1], c=predictions, edgecolors='k', cmap='Paired')
```
