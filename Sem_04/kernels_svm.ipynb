## Kernel interpretation

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from PIL import Image
from sklearn import svm
from sklearn.datasets import make_gaussian_quantiles
```

```
X, y = make_gaussian_quantiles(cov=2, n_samples = 250 , n_features = 2 , n_classes = 2 , random_state = 1)

X[y == 0] = X[y == 0] * 0.5

_, ax = plt.subplots(figsize=(6,5))
ax.scatter( X[:,0] , X[:,1] , c=y, cmap=plt.cm.Paired)
```

```
Z = X[:,0]**2 + X[:,1]**2
```

```
f, _ = plt.subplots(figsize=(6,5))
ax=f.add_subplot(projection="3d")
ax.view_init(10, 30)

ax.scatter3D( X[:,0] , X[:,1] , Z , c = y , s=10, cmap=plt.cm.Paired)
ax.set_zlim(0,5)

h_grid, v_grid = np.meshgrid(np.linspace(X[:,0].min(), X[:,0].max()), np.linspace(X[:,1].min(), X[:,1].max()))
z_grid = np.ones(h_grid.shape) * 1.5
ax.plot_surface(h_grid, v_grid, z_grid, alpha=0.5)

plt.tight_layout()
```

```
# Sve all figs
import os
os.makedirs("imgs/", exist_ok=True)

for angle in range(0, 91, 5):
  f, _ = plt.subplots(figsize=(6,5))
  ax = f.add_subplot(projection="3d")
  ax.view_init(angle, 30)

  ax.scatter3D( X[:,0] , X[:,1] , Z , c = y , s=10, cmap=plt.cm.Paired)
  ax.set_zlim(0,5)

  h_grid, v_grid = np.meshgrid(np.linspace(X[:,0].min(), X[:,0].max()), np.linspace(X[:,1].min(), X[:,1].max()))
  z_grid = np.ones(h_grid.shape) * 1.5
  ax.plot_surface(h_grid, v_grid, z_grid, alpha=0.5)

  plt.tight_layout()
  plt.savefig("imgs/{0}.png".format(angle))
  plt.close(), plt.clf()
```
